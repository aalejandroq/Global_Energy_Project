```python
a=RUS_df.reset_index()
a.head()
```

```python
# Date column only keeps year
a['Date'] = pd.DatetimeIndex(a['Date']).year
```

```python
a.head()
```

```python
#Reaname columns
a.columns =["Year","Agriculture","Mining","Manufacturing","Construction","Wholesale","Transport","Other","Total"]
a.head()
```

```python
# Add Country column
a['Country'] = tickers.Country[16]
a.head()
```

```python
#Create new df with desired column order
aa = a[ ['Country'] + [ col for col in a.columns if col != 'Country' ] ]
aa.head()
```

```python
# Calculate Total column
aa['Total'] = aa.loc[:,'Agriculture':'Other'].sum(axis=1)
```

```python
aa.head()
```

```python
aaa=aa.loc[:,"Agriculture":"Total"].div(aa["Total"], axis=0)*100
aaa.head()
```

```python
final = pd.DataFrame([aa['Country'],aa['Year'],aaa["Agriculture"],aaa["Mining"],aaa["Manufacturing"],aaa["Construction"],aaa["Wholesale"],aaa["Transport"],aaa["Other"]])
```
