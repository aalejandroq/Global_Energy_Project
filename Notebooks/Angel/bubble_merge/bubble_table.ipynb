```python
import pandas as pd
import sqlalchemy
import sqlite3 as sq
```

```python
# Creating table for the bubble plot
```

```python
# Read CSV file

emissions_file = 'CO2 emissions from fuel combus.csv'
tot_consumption_file = 'Total energy consumption.csv'
gdp_file = 'countries_gdp_location.csv'
```

```python
# Create a dataframe from csv file data

# Transform and clean emissions dataframe
emissions_df = pd.read_csv(emissions_file)
emissions_df = emissions_df.transpose()
emissions_df = emissions_df.reset_index()
emissions_df.columns = emissions_df.iloc[0]
emissions_df = emissions_df.reindex(emissions_df.index.drop(0))
emissions_df = emissions_df.rename(columns = {'Unnamed: 0':'Year'})
emissions_df = emissions_df.reindex(emissions_df.index.drop(29))
emissions_df = emissions_df.reindex(emissions_df.index.drop(30))
emissions_df.Year = emissions_df.Year.astype(int)

# Transform and clean consumption dataframe
tot_consumption_df = pd.read_csv(tot_consumption_file)
tot_consumption_df = tot_consumption_df.transpose()
tot_consumption_df = tot_consumption_df.reset_index()
tot_consumption_df.columns = tot_consumption_df.iloc[0]
tot_consumption_df = tot_consumption_df.reindex(tot_consumption_df.index.drop(0))
tot_consumption_df = tot_consumption_df.rename(columns = {'Unnamed: 0':'Year'})
tot_consumption_df = tot_consumption_df.reindex(tot_consumption_df.index.drop(29))
tot_consumption_df = tot_consumption_df.reindex(tot_consumption_df.index.drop(30))
tot_consumption_df.Year = tot_consumption_df.Year.astype(int)

# Create dataframe from GDP file, no transform needed
gdp_df = pd.read_csv(gdp_file)
```

```python
# Create list of countries for bubble plot
country_list = list(emissions_df.columns[1:])
country_list
```

```python
# Create list of years for bubble plot
year_list = list(emissions_df.Year)
year_list
```

```python
emissions = pd.melt(emissions_df, id_vars=['Year'], value_vars=country_list)
tot_consumption = pd.melt(tot_consumption_df, id_vars=['Year'], value_vars=country_list)
```

```python

emissions.columns = ['Year', 'Country', 'Emissions']
emissions.head()
```

```python
tot_consumption.columns = ['Year', 'Country', 'Consumption']
tot_consumption.head()
```

```python
# putting emissions and consumption data in 1 dataframe together
emissions['Consumption'] = tot_consumption['Consumption']
```

```python
emissions.head()
```

```python
# new dataframe with sorted data for emissions and consumption
emissions_consumption = emissions.sort_values(['Country','Year'])
```

```python
emissions_consumption.head()
```

```python
# From GDP table, grab the countries appearing the in energy tables
gdp_short_df = gdp_df[gdp_df['Country'].isin(country_list)]
gdp_short_df.shape
```

```python
# Likewise, filter the years in the dataframe to only show the years appearing in the energy tables
gdp_short_df = gdp_short_df[gdp_short_df['Year'].isin(year_list)]
gdp_short_df.shape
```

```python
gdp_short_df.head()
```

```python
# Add emissions and consumption data into one big dataframe
bubble_df = pd.merge(gdp_short_df, emissions_consumption, on=['Country', 'Year'])
```

```python
bubble_df.head()
```

```python
# New dataframe with only the columns needed for the bubble plot
bubble_final_df = bubble_df.filter(['Country','Year','Emissions','Continent','Consumption','Total'], axis=1)
```

```python
# renaming columns
bubble_final_df.columns = ['country','year','emissions','continent','consumption','gdp']
```

```python
bubble_final_df.tail()
```

```python
bubble_final_df.to_csv("bubble.csv")
```
